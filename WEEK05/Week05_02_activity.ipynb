```python
# Convert datetime column to Pandas datetime format
crime['datetimeocc'] = pd.to_datetime(crime['DATE OCC'], format='%m/%d/%Y %I:%M:%S %p')
crime['year'] = crime['datetimeocc'].dt.strftime('%Y')
crime['month'] = crime['datetimeocc'].dt.strftime('%B')
crime['day'] = crime['datetimeocc'].dt.strftime('%d')
crime['YM'] = crime['datetimeocc'].dt.to_period('M')
```
