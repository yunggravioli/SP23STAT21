```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Define set sizes
A_size = 20
B_size = 20
A_and_B_size = 10

# Create a Venn diagram with custom colors
venn_diagram = venn2(subsets=(A_size, B_size, A_and_B_size), 
                     set_labels=('A', 'B'),  alpha=0.5)

# Customize the appearance
venn_diagram.get_label_by_id('10').set_text('X')
venn_diagram.get_label_by_id('01').set_text('Y')
venn_diagram.get_label_by_id('11').set_text('X âˆ© Y')

intersection_patch = venn_diagram.get_patch_by_id('11')
intersection_patch.set_facecolor('purple')
intersection_patch.set_alpha(0.50)

A_patch = venn_diagram.get_patch_by_id('10')
A_patch.set_facecolor('white')
A_patch.set_alpha(0.50)

B_patch = venn_diagram.get_patch_by_id('01')
B_patch.set_facecolor('white')
B_patch.set_alpha(1.0)

# Add black borders to sets A and B
for set_id in ['10', '01']:
    patch = venn_diagram.get_patch_by_id(set_id)
    patch.set_edgecolor('black')
    patch.set_linewidth(2)


# Display the Venn diagram
# plt.show()
plt.savefig('inner.png')
```
